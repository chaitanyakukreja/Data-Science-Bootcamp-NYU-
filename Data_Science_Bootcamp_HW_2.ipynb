```python
# HW QUESTIONS (NUMPY)

import numpy as np
```

```python
#Define two custom numpy arrays A and B
A = np.array([[1, 5, 9],
              [4, 7, 2]])
B = np.array([[3, 5, 8],
              [1, 7, 6]])

print("Array A:\n", A)
print("\nArray B:\n", B)
```

```python
#Stack A and B vertically and horizontally
vertical_stack = np.vstack((A, B))
horizontal_stack = np.hstack((A, B))

print("\nVertically stacked (A over B):\n", vertical_stack)
print("\nHorizontally stacked (A beside B):\n", horizontal_stack)
```

```python
#Find common elements between A and B (intersection)
common_elements = np.intersect1d(A, B)
print("\nCommon elements between A and B:\n", common_elements)
```

```python
#Extract all numbers from A which are between 5 and 10
filtered_A = A[(A >= 5) & (A <= 10)]
print("\nNumbers in A between 5 and 10:\n", filtered_A)
```

```python
#Filter rows of iris_2d with petallength > 1.5 and sepallength < 5.0
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

filtered_iris = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]
print("\nFiltered iris rows (petal length > 1.5 and sepal length < 5.0):\n", filtered_iris)
```

```python
# HW QUESTIONS (PANDAS)
import pandas as pd
```

```python
#Filter 'Manufacturer', 'Model', and 'Type' for every 20th row starting from row 0
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
filtered_df = df.loc[::20, ['Manufacturer', 'Model', 'Type']]
print("\nEvery 20th row (Manufacturer, Model, Type):\n", filtered_df)
```

```python
#Replace missing values in Min.Price and Max.Price columns with their respective mean
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df['Min.Price'].fillna(df['Min.Price'].mean(), inplace=True)
df['Max.Price'].fillna(df['Max.Price'].mean(), inplace=True)
print("\nMissing values in Min.Price and Max.Price replaced with their mean:")
print(df[['Min.Price', 'Max.Price']].head())
```

```python
#Get rows of a dataframe with row sum > 100
df_random = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4),
                         columns=['A', 'B', 'C', 'D'])
filtered_rows = df_random[df_random.sum(axis=1) > 100]

print("\nRandom DataFrame:\n", df_random)
print("\nRows with row sum > 100:\n", filtered_rows)
```
