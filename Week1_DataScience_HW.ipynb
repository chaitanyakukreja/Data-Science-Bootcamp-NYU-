```python
# 1. Count vowels in a word
word = input("Enter a word: ")
vowels = "aeiouAEIOU"
count = 0
for char in word:
    if char in vowels:
        count += 1
print(f"Number of vowels in '{word}':", count)
```

```python
# 2. Print animal names in all caps
animals = ['tiger', 'elephant', 'monkey', 'zebra', 'panther']
print("Animals in all caps:")
for animal in animals:
    print(animal.upper())
```

```python
# 3. Print numbers from 1 to 20 and check if odd or even
for num in range(1, 21):
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
# 4. Check if a string is a palindrome
text = input("Enter a word to check palindrome: ")
if text.lower() == text[::-1].lower():
    print(f"'{text}' is a palindrome.")
else:
    print(f"'{text}' is not a palindrome.")
```

```python
# 5. Function to sum two integers
def sum_of_integers(a, b):
    return a + b

# (optional part â€” getting input from user)
a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))
print("The sum is:", sum_of_integers(a, b))
```
